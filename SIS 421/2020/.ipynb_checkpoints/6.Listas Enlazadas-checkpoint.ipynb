```python
class ListaEnlazada():
    
    def __init__(self):#Recordar que None equivale a null en otros Lenguajes
        self.first = None
        self.last = None 
    
    #Funcion para Ver Si el nodo esta Vacio
    def Void(self):
        return self.first == None
    
    #Funcion para insertar un elemento al Principio de la Lista
    def InsertFirst(self,Data):
        if self.Void() == True:
            self.first = self.last = Nodo(Data)
        else:
            Aux = Nodo(Data)
            Aux.Next = self.first
            self.first = Aux 
            
    #Funcion para insertar un Elemento despues del Existente
    def InsertLast(self,Data):
        if self.Void() == True:
            self.first = self.last = Nodo(Data)
        else:
            Aux = self.last
            self.last = Aux.Next = Nodo(Data)
            
    #Funcion para Eliminar el primer Elemento 
    def DeleteFirst(self):
        self.first = self.first.Next
        
    
    #Funcion para Eliminar el ultimo Elemento 
    def DeleteLast(self):
        Aux = self.first
        while Aux.Next != self.last:
            Aux = Aux.Next
        Aux.Next = None
        self.last = Aux

    #Funcion para Recorrer la Lista Enlazada
    def Travel(self):
        Aux = self.first
        while Aux != None:
            print(Aux.Data)
            Aux = Aux.Next
        
```

```python
ListaEnlazada = ListaEnlazada();

ListaEnlazada.InsertLast('POO');
ListaEnlazada.InsertLast('Encapsulamiento');
ListaEnlazada.InsertLast('Polimorfismo');
ListaEnlazada.InsertLast('Herencia');
ListaEnlazada.InsertLast('Abstraccion');

ListaEnlazada.InsertFirst('Paradigmas de programacion:')
ListaEnlazada.InsertFirst('***************444**************************')

ListaEnlazada.Travel();

print('********************0******************')
print(ListaEnlazada.first.Data)
print(ListaEnlazada.last.Data)
print('**************************************')

ListaEnlazada.DeleteFirst()
ListaEnlazada.DeleteFirst()
#ListaEnlazada.DeleteLast()
#ListaEnlazada.DeleteLast()
#ListaEnlazada.DeleteLast()
ListaEnlazada.Travel();
```

# Listas Doblemente Enlazadas

```python
class Nodo():
    
    def __init__(self,Data):
        self.Data = Data
        self.Next = None
        self.Previous = None
        
        
class ListaDoblementeEnlazada():        
    
    def __init__(self):
        self.First = None
        self.Last = None
        self.Size = 0
    
    def Void(self):
        return self.First == None
    
    def InsertLast(self,Data):
        if self.Void() == True:
            self.First = self.Last = Nodo(Data)
        else:
            Aux = self.Last
            self.Last = Aux.Next = Nodo(Data)
            self.Last.Previous = Aux
        self.Size += 1
        
    def InsertFirst(self,Data):
        if self.Void() == True:
            self.First = self.Last = Nodo(Data)
        else:
            Aux = Nodo(Data)
            Aux.Next = self.First
            self.First.Previous =  Aux
            self.First = Aux
        self.Size += 1
        
    def DeleteFirst(self):
        if self.Void() == True:
            print('La lista esta Vacia')
        elif self.First.Next == None :
            Aux = self.First
            self.First = self.Last = None
            self.Size = 0 
        else: 
            self.First = self.First.Next
            self.First.Previous = None
            self.Size -= 1
        
    def DeleteLast(self):
        if self.Void() == True:
            print('La lista esta Vacia')
        elif self.First.Next == None :
            self.First = self.Last = None
            self.Size = 0
        else:
            self.Last = self.Last.Previous
            self.Last.Next = None
            self.Size -= 1
        
    def TravelStart(self):
        Aux = self.First
        while Aux != None:
            print(Aux.Data)
            Aux = Aux.Next
            
    def TravelEnd(self):
        Aux = self.Last
        while Aux != None:
            print(Aux.Data)
            Aux = Aux.Previous
            
    def SizeList(self):
        return print(self.Size)
```

```python

Lista = ListaDoblementeEnlazada()

Lista.InsertFirst(10)
Lista.InsertFirst(-9)
Lista.InsertFirst(-9)
Lista.InsertFirst(-9)
Lista.InsertFirst(-9)
Lista.InsertFirst(-10)
Lista.InsertFirst('El Principio es el Fin')
Lista.InsertLast('El Fin es el Principio')
Lista.SizeList()

print('-------------------------')
Lista.TravelStart()
print('-------------------------')
Lista.TravelEnd()

print('masacre')
Lista.DeleteFirst()
Lista.DeleteLast()
Lista.TravelStart()
```

# Lista Circular Simple

## Ejemplo 0 

```python
class Nodo():#Clase Nodo para Instanciar un Objeto Nodo
    
    #Funcion Constructor
    def __init__(self,Data):
        self.Data = Data
        self.Next = None

class ListaCircularSimple(): # Clase ListaCircularSimple para Instanciar un Objeto ListaCircularSimple
    
    #Funcion Constructor
    def __init__(self):
        self.First = None
        self.Last = None
    
    #Funcion para Verificar si la Lista esta vacia
    def Void(self):
        return self.First == None
    
    #Funcion para insertar un Elemento al Principio de La Lista
    def InsertFirst(self,Data):
        if self.Void() == True:
            self.First = self.Last = Nodo(Data)
            self.Last.Next = self.First
        else:
            Aux = Nodo(Data)
            Aux.Next = self.First  
            self.First = Aux
            self.Last.Next = self.First

    #Funcion para insertar un Elemento al Final del Lista
    def InsertLast(self,Data):
        if self.Void() == True: 
            self.First = self.Last = Nodo(Data)
            self.Last.Next = self.First
        else:
            Aux = self.Last
            print('Aux : ',Aux.Data)
            self.Last = Aux.Next = Nodo(Data)
            self.Last.Next = self.First
            
    #Funcion para eliminar un Elemento al Principio
    def DeleteFirst(self):
        if self.Void() == True:
            print('Lista esta Vacia')
        elif self.First == self.Last :
            self.First = self.Last = None
        else:
            self.First = self.First.Next
            self.Last.Next = self.First

            
    #Funcion para eliminar un Elemento al Final
    def DeleteLast(self):
        if self.Void() == True:
            print('Lista Vacia')
        elif self.First == self.Last:
            self.First = self.Last = None
        else:
            Aux = self.First
            while Aux.Next != self.Last:
                Aux = Aux.Next
            Aux.Next = self.First
            self.Last = Aux

    #Funcion para Mostrar ListaE Circular
    def Travel(self):
        Aux = self.First
        while Aux:
            print(Aux.Data)
            Aux = Aux.Next
            if Aux == self.First:
                break
        
```

```python
Lista = ListaCircularSimple()

Lista.InsertLast(10)
Lista.InsertLast(9)
Lista.InsertLast(8)
Lista.InsertLast(7)
Lista.InsertFirst(100)
Lista.InsertFirst(200)

Lista.Travel()

print('*****************************')
Lista.DeleteFirst()
Lista.DeleteLast()
Lista.Travel()
```

```python
class Nodo():
    
    def __init__(self,Data):
        self.Data = Data
        self.Next = None
        self.Previous = None
        
class ListaCircularDoblementeEnlazada():
    
    def __init__(self):
        self.First = None
        self.Last = None

    def Void(self):
        if self.First == None:
            return True
        else: 
            return False
    
    def InsertFirst(self,Data):
        if self.Void() == True:
            self.First = self.Last = Nodo(Data)
        else:
            Aux = Nodo(Data)
            Aux.Next = self.First
            self.First.Previous = Aux
            self.First = Aux
        self.__Unir()
        
    def InsertLast(self,Data):
        if self.Void() == True:
            self.First = self.Last = Nodo(Data)
        else:
            Aux = self.Last # void 0 
            self.Last = Aux.Next = Nodo(Data)# 10
            self.Last.Previous = Aux # 9  = 9  x = 9 
        self.__Unir()

    def DeleteFirst(self):
        if self.Void() == True:
            print('Lista esta Vacia no se Puede Elminar Elementos')
        elif self.First == self.Last :
            self.First = self.Last = None
        else:
            self.First = self.First.Next
        self.__Unir()
            
            
    def DeleteLast(self):
        if self.Void() == True:
            print('Lista esta Vacia no se Puede Elminar Elementos')
        elif self.First == self.Last :
            self.First = self.Last = None
        else:
            self.Last = self.Last.Previous
        self.__Unir()

    def Search(self,Value):
        if self.Void() == True:
            print('Lista esta Vacia no se Puede Buscar Elementos')
        Aux = self.First
        while Aux:
            #'DatoActual:',Aux.Data,'=Value',Value)
            if Aux.Data == Value:
                print('Dato Encontrado')
                break
            else:
                Aux = Aux.Next
                if Aux == self.First:
                    print('El Dato no se Encuentra en la Estructura')
                    break
                    
    def __Unir(self):
        if self.First != None:
            self.First.Previous = self.Last
            self.Last.Next = self.First
        
    def TravelStartToEnd(self):
        Aux = self.First
        while Aux:
            print(Aux.Data)
            Aux = Aux.Next
            
            if Aux == self.First:
                break
            
    def TravelEndToStart(self):
        Aux = self.Last
        while Aux:
            print(Aux.Data)
            Aux = Aux.Previous
            
            if Aux == self.Last:
                break

        
```
