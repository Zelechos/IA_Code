## 1.- Algoritmo de Ordenamiento por Insercion
El ordenamiento por inserción es una manera muy natural de ordenar para un ser humano, y puede usarse fácilmente para ordenar un mazo de cartas numeradas en forma arbitraria.

La idea de este algoritmo de ordenación consiste en ir insertando un elemento de la lista ó un arreglo en la parte ordenada de la misma, asumiendo que el primer elemento es la parte ordenada, el algoritmo ira comparando un elemento de la parte desordenada de la lista con los elementos de la parte ordenada, insertando el elemento en la posición correcta dentro de la parte ordenada, y así sucesivamente hasta obtener la lista ordenada.

```python
Vector = [4,6,0,7,2,5,1,3,10,9,8]
def Insertion_sort(Vector):
    for i in range(1,len(Vector)):
        actual = Vector[i]
        j = i
        #Desplazamiento de los elementos de la matriz 
        while j>0 and Vector[j-1]>actual:
            Vector[j]=Vector[j-1]
            j = j-1
        #insertar el elemento en su lugar
        Vector[j]=actual
    return print('Lista Ordenada : \n',Vector)        

Insertion_sort(Vector)
```

```python
lista = [4,5,3,8,2,0,9]
print('Lista Actual:\n',lista)
def Asendente(lista):
    
    for index in range(1,len(lista)):
        actual = lista[index]
        antecesor = index

        while antecesor > 0 and lista[antecesor-1] > actual:
            lista[antecesor] = lista[antecesor-1]
            antecesor -= 1
        lista[antecesor] = actual
    
    return print('Lista Ordenada Asendente: \n',lista)                

def Desendente(lista):
    
    for index in range(1,len(lista)):
        actual = lista[index]
        antecesor = index
        
        while antecesor > 0 and lista[antecesor-1] < actual:
            lista[antecesor] = lista[antecesor-1]
            antecesor -= 1
        lista[antecesor] = actual
    return print('Lista Ordenada Desendente: \n',lista)                
#actual 

Asendente(lista) 
Desendente(lista)
```

## 5.-Algoritmo de Ordenamiento Rapido (Quick Sort)

```python
vector = [1,3,2,5,4]

print(vector[:-1])#todos los valores del la lista menos el ultimo

def quick_sort(vector):
    if not vector:
        return []
    else:
        pivote = vector[-1]
        menor = [x for x in vector     if x <  pivote]
        print('menor : ',menor)
        mas_grande = [x for x in vector[:-1] if x >= pivote]
        print('mas_grande : ',mas_grande)
        return quick_sort(menor) + [pivote] + quick_sort(mas_grande)

quick_sort(vector)
```

```python
lista = [3,8,4,6,9,3,1,4,5,76,87,9,100,-11,-90,0]

def Desendente(lista):
    if not lista:
        return []
    else:
        Pivote = lista[-1]
        menor = [x for x in lista   if x < Pivote]
        mayor = [x for x in lista[:-1]  if x >= Pivote]
        return Desendente(mayor) + [Pivote] + Desendente(menor)

Desendente(lista)
```
